## **Outlier's Handling**

```python
# Sanity Check (Outliers)

# Fungsi untuk mendeteksi outlier menggunakan IQR
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (series < lower_bound) | (series > upper_bound)
    return outliers

# Mendeteksi outlier untuk setiap kolom
outliers = data.apply(detect_outliers_iqr)

# Tampilkan outlier untuk setiap kolom
print("Outlier for each column:")
print(outliers)
```

## **Relational Checking antar Segmentasi dengan Target Column**

### **Baseline Value Segment & Fetal Health Segment Checking**

```python
from scipy.stats import chi2_contingency

# Menghitung tabel kontingensi antara kategori segmentasi dan kesehatan janin
contingency_table = pd.crosstab(index=data['baseline_segment'], columns=data['fetal_health'])

# Melakukan uji chi-square
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Menampilkan hasil uji chi-square
print("Hasil Uji Chi-Square:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")

# Interpretasi hasil uji chi-square
alpha = 0.05
if p < alpha:
    print("Terdapat hubungan yang signifikan antara segmentasi baseline dan kesehatan janin.")
else:
    print("Tidak terdapat hubungan yang signifikan antara segmentasi baseline dan kesehatan janin.")
```

## Ridge Regression

```python
# Data Splitting
from sklearn.model_selection import train_test_split

X = data[['fetal_movement','abnormal_short_term_variability','percentage_of_time_with_abnormal_long_term_variability','histogram_width','histogram_max','histogram_mode','histogram_mean','histogram_variance']].to_numpy()
y = data['fetal_health'].to_numpy()
y = y.reshape(len(y),)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20 , random_state=42) #To make sure the train and test split is 8/2
```

```python
# define ridge regression model
from sklearn.linear_model import Ridge

ridge_reg = Ridge(random_state=42)
```

```python
# tune lambda (denote as 'alpha' in sklearn)
# using GridSearchCV
from sklearn.model_selection import GridSearchCV

# hyperparameter values we want to tune
parameters = {
    'alpha': (0.000001,0.00001,0.0001,0.001,
              0.01, 0.1, 1, 5, 10, 20)
}

# the tuning
ridge_reg_gridcv = GridSearchCV(ridge_reg, parameters, cv=5,
                                scoring='neg_root_mean_squared_error')
ridge_reg_gridcv.fit(X_train, y_train)
```

```python
# Function untuk plot learning curve model-model kita
def plot_learning_curve(estimator, title, X, y, ylim=None,
                        n_jobs=None, train_sizes=np.linspace(0.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.grid(True)
    plt.title(title,
              fontsize=24,
              fontweight='bold')
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=10, n_jobs=n_jobs, train_sizes=train_sizes,
        scoring='accuracy', shuffle=True, random_state=42)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="b")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="b",
             label="Cross-validation score")

    plt.legend(loc="best")
    plt.tight_layout()
    return plt
```

Dokter atau ahli gizi dapat memberikan saran yang sesuai dengan kebutuhan individu ibu hamil.

**Hindari Zat Berbahaya:** Menghindari zat-zat yang berpotensi berbahaya seperti alkohol, rokok, dan obat-obatan terlarang sangat penting untuk mencegah masalah perkembangan janin. Konsultasikan dengan dokter tentang penggunaan obat-obatan tertentu, termasuk obat resep dan obat bebas, selama kehamilan.

**Penghindaran Paparan Racun:** Usahakan untuk menghindari paparan terhadap bahan kimia berbahaya seperti pestisida, bahan pembersih yang keras, dan zat-zat kimia industri lainnya yang dapat membahayakan perkembangan janin.

**Olahraga yang Tepat:** Melakukan olahraga ringan dan aktivitas fisik yang sehat dapat membantu menjaga kesehatan ibu dan janin. Berbicaralah dengan dokter untuk menentukan jenis dan tingkat aktivitas fisik yang sesuai selama kehamilan.

**Istirahat yang Cukup**: Istirahat yang cukup juga penting selama kehamilan untuk membiarkan tubuh beristirahat dan memulihkan diri. Usahakan untuk tidur cukup setiap malam dan beristirahat secara teratur selama hari.

**Manajemen Stres**: Stres yang berlebihan dapat berdampak negatif pada kesehatan janin. Temukan cara untuk mengelola stres seperti meditasi, yoga, atau aktivitas santai lainnya yang dapat membantu menjaga keseimbangan emosional.

**Pantau Perkembangan Janin**: Selain pemantauan detak jantung janin, pemantauan pertumbuhan janin secara berkala juga penting. Ini dapat dilakukan melalui ultrasonografi dan pemeriksaan fisik oleh dokter atau bidan.

**Konsultasi dengan Dokter**: Jika ada kekhawatiran atau pertanyaan tentang kesehatan janin, penting untuk segera berkonsultasi dengan dokter atau tenaga medis yang berkualifikasi. Mereka dapat memberikan nasihat dan perawatan yang tepat berdasarkan kondisi spesifik ibu hamil dan janinnya.

Menjalani gaya hidup sehat dan mengikuti perawatan prenatal yang tepat dapat membantu memastikan kesehatan janin selama masa kehamilan.
